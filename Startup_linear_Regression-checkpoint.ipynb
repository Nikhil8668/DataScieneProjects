```python
## Step 4:-Checking the outliers
plt.figure(figsize=(20,25),facecolor='white')
plotnumber=1

for column in data.drop('Profit',axis=1):
    if plotnumber<4:
        ax=plt.subplot(2,2,plotnumber)
        sns.boxplot(data[column])
        plt.xlabel(column,fontsize=20)
        plt.ylabel('Count',fontsize=20)
    plotnumber+=1
plt.tight_layout()
```

```python
#drop the unnamed column
data.drop('State',axis = 1,inplace=True)
```

```python
data.corr()
```

```python
#Checking the correlation
sns.heatmap(data.corr(),annot=True)
```

```python
sns.boxplot(x=data['Profit'])
plt.show()
```

```python
#Outlier Handilng
q1 = data['Profit'].quantile(0.25)
q3 = data['Profit'].quantile(0.75)
print(q1)
print(q3)
```

```python
iqr = q3 - q1
iqr
```

```python
min1 = q1 - 1.5*iqr
min1
```

```python
data.loc[data['Profit']<min1]
```

```python
data.loc[data['Profit']<min1]=data['Profit'].median()
```

```python
data.loc[data['Profit']<min1]
```

```python
sns.boxplot(x=data['Profit'],data=data)
plt.show()
```

```python
#Splitting the data into x and y
x=data[['R&D Spend','Administration','Marketing Spend']]
y=data.Profit
```

```python
x
```

```python
y
```

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.30,random_state=4)
```

```python
x_train
```

```python
x_test.shape
```

```python
y_train
```

```python
y_test.shape
```

# Model Building

```python
from sklearn.linear_model import LinearRegression
model = LinearRegression()
model.fit(x_train,y_train)
```

```python
model
```

```python
#Make PRedictions
y_predict = model.predict(x_test)
y_predict
```

```python
#Actual Values
y_test.head()
```

# Evaluation Model

```python
from sklearn.metrics import r2_score,mean_squared_error,mean_absolute_error
```

```python
#MSE Score
mse = mean_squared_error(y_test,y_predict)
mse
```

```python
mae = mean_absolute_error(y_test,y_predict)
mae
```

```python
rmse=np.sqrt(mse)
rmse
```

```python
r2score=r2_score(y_test,y_predict)
r2score
```
